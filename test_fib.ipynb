contract get_fib_n(n:int){
    let last_fib_calculated: int = 0;

    entry get_fib_n(n: int){
        let result: int = fib(n);
        last_fib_calculated = result;
    }

    func fib(n: int) : int{
        if (n <= 1) {
            return n;
        }
        else {
            return fib(n-1) + fib(n-2);
        }
    }
}
